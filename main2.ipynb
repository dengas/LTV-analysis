```python
# Задание:

# Ты аналитик в компании, занимающейся онлайн-продажей электроники. Твоя задача - построить анализ клиентских когорт, чтобы определить, как изменяется средний доход с клиента (LTV) в зависимости от времени его первого заказа.

# Описание данных:
# orders - таблица с информацией о заказах: order_id - уникальный идентификатор заказа customer_id - уникальный идентификатор клиента
# order_date - дата, когда заказ был создан order_amount - сумма заказа platform - платформа, через которую был сделан заказ (например, "mobile", "desktop") category - категория товара, который был заказан (например, "smartphones", "laptops")

# Задача: Напишите код, который выполнит следующие действия:

# Построй когортный анализ: выдели клиентов, совершившие первую покупку в март 2020, и вычисли размер каждой когорты. Рассчитай средний доход с клиента для каждой когорты за 10 дней с момента их первой покупки в январе 2020. Отобрази результаты по месяцам первых покупок, категориям товаров и платформам. Ожидаемый результат: Вывод должен включать следующие поля:

# Дата когорты (дата первой покупки клиента)
# Платформа 
# Категория товара 
# Размер когорты 
# Средний доход с клиента
```

```python
import pandas as pd
import numpy as np

orders2 = pd.read_csv('orders.csv')
orders2['order_date'] = pd.to_datetime(orders2['order_date'], errors='coerce')
```

```python
orders2.dtypes
```

```python
orders2[orders2['order_date'] == '2020-03-01'].count()
```

```python
first_order_df = orders2.groupby(['customer_id'], as_index=False).agg({'order_date': 'min'}).rename(columns={'order_date' : 'first_order_date'})
first_order_df
```

```python
orders2 = orders2.merge(first_order_df, on='customer_id')
```

```python
orders2['delta_days'] = (orders2['order_date'] - orders2['first_order_date']).dt.days
orders_first_10_days = orders2[orders2['delta_days'] < 10]
orders_first_10_days

orders_first_10_days[orders_first_10_days['customer_id'] == 3]
```
